```python
import folium
import os
os.listdir()
```

```python
map = folium.Map(location = [38.5, -99], zoom_start = 6)
map.save("Map1.html")
```

```python
map2 = folium.Map(location = [38.5, -99], zoom_start = 6, tiles = "Mapbox Bright")

fg = folium.FeatureGroup(name = "My map")
fg.add_child(folium.Marker(location = [38.2, -99.1], popup = "Hi, I'm a Marker", icon = folium.Icon(color="green")))
map2.add_child(fg)

map2.save("Map2.html")
```
