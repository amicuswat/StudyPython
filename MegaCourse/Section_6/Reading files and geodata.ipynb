```python
df10 = pandas.read_csv("supermarkets.csv")
df10
```

```python
df10["Address"] = df10["Address"] + ", " + df10["City"] + ", " + df10["State"] + ", " + df10["Country"]
df10
```

```python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
n = nom.geocode("3666 21st St, San Francisco, CA 94114, USA")
n.longitude
```

```python
n.latitude
```

```python
df10["Coordinates"] = df10["Address"].apply(nom.geocode)
df10
```

```python
df10.Coordinates[0]
```

```python
df10["Latitude"] = df10["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df10["Longitude"] = df10["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df10
```
